```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Lennard-Jones potential function
def lennard_jones(r, sigma, E_0):
    return 4 * E_0 * ((sigma / r) ** 12 - (sigma / r) ** 6)

# Generate x values
r = np.linspace(0.001, 3, 100)

# Define different values of sigma and E_0
sigma_values = [1, 1.5, 2]
E_0_values = [1]

# Plot the Lennard-Jones potential for different values of sigma and E_0
for sigma in sigma_values:
    for E_0 in E_0_values:
        y = lennard_jones(r, sigma, E_0)
        plt.plot(r, y, label=f"sigma={sigma}, E_0={E_0}")

plt.xlabel('r')
plt.ylabel('Potential')
plt.title('Lennard-Jones Potential for Different Values of sigma and E_0')
plt.ylim(-1, 1)
plt.legend()
plt.show()
```

```python
# Define the smooth linear Lennard-Jones potential function
def smooth_linear_lennard_jones(r, sigma, epsilon, rc):
    smooth_r = np.where(r < rc, r, rc)
    return 4 * epsilon * ((sigma / smooth_r) ** 12 - (sigma / smooth_r) ** 6)

# Generate x values
r = np.linspace(0.001, 3, 100)

# Define the parameters
sigma_values = [1, 1.5, 2]
epsilon = 1e-1
rc = 2

# Plot the smooth linear Lennard-Jones potential for different values of sigma
for sigma in sigma_values:
    y = smooth_linear_lennard_jones(r, sigma, epsilon, rc)
    plt.plot(r, y, label=f"sigma={sigma}")

plt.xlabel('r')
plt.ylabel('Potential')
plt.title('Smooth Linear Lennard-Jones Potential for Different Values of sigma')
plt.ylim(-1, 1)
plt.legend()
plt.show()
```

```python
# Define the Morse potential function
def morse_potential(r, D, a, r_0):
    return D * (np.exp(-2 * a * (r - r_0)) - 2 * np.exp(-a * (r - r_0)))

# Generate x values
r = np.linspace(0.001, 3, 100)

# Define different values of r_0 and D
r_0_values = [0.5, 1, 2]
D_values = [2]

# Define the parameter a
a = 1

# Plot the Morse potential for different values of r_0 and D
for r_0 in r_0_values:
    for D in D_values:
        y = morse_potential(r, D, a, r_0)
        plt.plot(r, y, label=f"r_0={r_0}, D={D}")

plt.xlabel('r')
plt.ylabel('Potential')
plt.title('Morse Potential for Different Values of r_0 and D')
plt.ylim(-2, 1)
plt.legend()
plt.show()
```

```python
def smooth_linear_morse(r,D,alpha,r_0):
    return D*(np.exp(-2*alpha*(r-r_0))-2*np.exp(-alpha*(r-r_0))**2)
    
r = np.linspace(0.001, 3, 100)
D = 1e-1
alpha = 1
r_0 = 1
y = smooth_linear_morse(r,D,alpha,r_0)
plt.plot(r,y)
plt.xlabel('r')
plt.ylabel('Potential')
plt.title('Smooth Linear Morse Potential')
plt.show()
```
