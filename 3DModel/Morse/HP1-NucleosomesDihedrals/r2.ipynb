```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the .dat file into a pandas DataFrame
df = pd.read_csv('r2.dat', comment='#', delim_whitespace=True, names=['TimeStep', 'R_g'])

# Plot c_R2[0] against TimeStep
plt.plot(df['TimeStep'], df['R_g'])
plt.xlabel('TimeStep')
plt.ylabel('$R_g$')
plt.title('Plot of $R_g$ vs TimeStep')
plt.grid(True)
plt.show()
```
