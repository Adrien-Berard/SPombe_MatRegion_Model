```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/Users/adrienberard/Documents/GitHub/InternshipKU/4statesModelPhospo.csv")
```

```python
plt.plot("Time Steps","Acetylated Histone Count",data=df,label="Acetylated Histone")
plt.plot("Time Steps","Methylated Histone Count",data=df,label="Methylated Histone")
plt.plot("Time Steps","Phosphorylated Histone Count",data=df,label="Phosphorylated Histone")
plt.plot("Time Steps","Unmodified Histone Count",data=df,label="Unmodified Histone")
plt.legend()
plt.xlabel("Time Steps")
plt.ylabel("Histone Count")
plt.title("Histone Modification")

plt.show()
```

```python
Noisy_cum_sum = np.cumsum(df["Noisy Changes Count"])
Enzymes_cum_sum = np.cumsum(df["Enzyme Changes Count"])
time = df["Time Steps"]
plt.plot(time,Noisy_cum_sum,label="Noisy Changes")
plt.plot(time,Enzymes_cum_sum,label="Enzymes Changes")
plt.legend()
plt.plot()
```
